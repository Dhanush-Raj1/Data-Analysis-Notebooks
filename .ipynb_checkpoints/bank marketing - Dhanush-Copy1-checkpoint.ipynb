# Objective :
   The main objective of the research work is to find the best suitable way for marketing term deposits of a bank. We are required to suggest ways to improve the marketing campaign of a bank to increase the term deposit subscriptions.

# Hypothesis:

### Null hypothesis:
   There is a relationship and direct impact between marketing campaign and subscription of term deposits.
  
### Alternative hypothesis:
   There is no relationship and impact between marketing campaign and subscription of term deposits.

## Overview of the Data (secondary data)
       
   - Our data set totally contains 21 columns .
   - We have clients data
        - age
        - education
        - job
        - default    (has credit loan or not)
        - housing    (has house loan or not)
        - contact
    
   -  We also have other attributes such as
         - pdays     (number of days passed after the client was last contacted from a previous campaign)
         - poutcome  (outcome of the previous marketing campaign)
   
   - And other economic attributes such as
        - emp.var.rate (employee varation rate)
        - cons.price.idx (consumer price index)
        - cons.confi.idx (consumer confidence index)
        
   - **y column is dependent variable (subscription of term deposit or not).**
   - **Age, education, job, default, housing, contact,.... is independent variable.**

# Vizualizing the DISTRIBUTION of variable 

```python
sns.pairplot(bank_data)
```

## Observation
  **Distribution is the graphical representation of a data to understand the overall info of the data.**  
  **Our motive is to find wheather our data is normally distributed or not. In a normal distribution, the data is arranged in a way that most of the values form a cluster in the middle and taper off in a symmetric fashion towards either extreme. It is also called a Gaussian distribution. It appears as a bell curve when shown graphically. In a standard normal distribution, the mean is zero, and the standard deviation takes the value of 1 along with a zero skew. The mean, median and mode are all the same in a normal distribution**  
   
   - From the above vizualizations we can interpret that our data is not normally distributed. It is because our data is realtime time data collected by means of surveys and through customer's data.
   - Realtime data usually contains many noices and errors and outliers it is one of the factor for defining the distribution of our data.
   - Another reason is the timeline of our data. It may have took longer duration to collect data.

```python
### predicting 
pred=log_rig.predict(x_train_new)  
predd=log_rig.predict(x_test_new)
```

```python
from sklearn.metrics import r2_score
train_r2_score=r2_score(y_train,pred)
print('training coefficient of detemination:',train_r2_score)
```
